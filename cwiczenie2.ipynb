```julia
using JuMP
using Ipopt

m = Model(solver = IpoptSolver())

demand = 200
# demand = 300

@variable(m, 70 <= A <= 200)
@variable(m, 80 <= B <= 250)
@constraint(m, A+B == demand)

@objective(m, Min, 140*A+0.8*B*B+0.5*B)

print(m)

status = solve(m)
```
